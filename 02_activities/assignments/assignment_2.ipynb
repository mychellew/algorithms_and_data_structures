```python
from collections import deque

class TreeNode:
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

def is_duplicate(root: TreeNode) -> int:
    if not root:
        return -1  # No tree, no duplicates (-1)

    seen = set() # Set to store the values of the nodes already seen
    queue = deque([root])  # BFS queue. What the queue stores is the nodes, not the values

    while queue:
        node = queue.popleft()

        # If the value is already seen, return it as the closest duplicate
        if node.val in seen:
            return node.val
        seen.add(node.val)

        # Process children nodes
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

    return -1  # No duplicates found
```


-   Explain why their solution works in your own words.


```python
'''
The solution that Carlos developed works as it utliizes BFS (level-order
traversal) to check the values closest to the root and continue to iterate 
through the tree. Firstly, in the best-case scenario, it will return the value
-1 immediately if the input provided is not a tree. Otherwise if a valid input
is provided to the function, it will iterate through the tree level by level 
and store the iteraed values in a set. Each time the function checks the next 
level, it will also check the list of seen values. Once it has identified one 
that matches, it will return that value as the duplicate and stop iterating 
through the rest of the function. If it reaches the end of the tree and there 
are not duplicates then it will return a value of -1
'''
```


-   Explain the problem’s time and space complexity in your own words.


```python
'''
Time Complexity: O(n)
When discussing Big-O, the worst case scenario for this algorithm is O(n).
If there are no duplicates within the Tree, the algorithm will have 
"checked" through each value of the tree in order to provide a value of -1

Space Complexity: O(n)
The space complexity for this algorithm is O(n), in the scenario where there
are no duplicate values, the seen set of values will store all of the unique
values (n). In the worst case scenario, the queue stores all of the nodes
present in the tree which is O(n) if each node in a tree does not have 
both left and right children.

'''
```
